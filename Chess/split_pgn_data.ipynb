```python
import os
import chess.pgn

def split_pgn(input_pgn_file, output_directory):
    with open(input_pgn_file, 'r') as pgn_file:
        pgn = chess.pgn.read_game(pgn_file)
        game_number = 1

        while pgn:
            output_pgn_file = os.path.join(output_directory, f'game_{game_number}.pgn')
            with open(output_pgn_file, 'w') as output_file:
                output_file.write(str(pgn))
            print(f'Saved game {game_number} to {output_pgn_file}')

            pgn = chess.pgn.read_game(pgn_file)
            game_number += 1

if __name__ == "__main__":
    input_pgn_file = 'D:/AI/Chess/ChessApp/Chess/raw_pgn/Abdusattorov.pgn'  # Replace with the path to your input PGN file
    output_directory = 'pgn_data/abdusattorov'  # Specify the output directory for split PGN files
    
    # Create the output directory if it doesn't exist
    if not os.path.exists(output_directory):
        os.makedirs(output_directory)

    split_pgn(input_pgn_file, output_directory)
```
